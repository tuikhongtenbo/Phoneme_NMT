```python
# if word in train/dev/test -> return indexes of that line and add to arrays
def index_non_alpha(dataset: list, non_alpha_non_arr: list) -> set:
    # Đưa list từ cần tìm thành set để tra cứu nhanh
    target_words = set(non_alpha_non_arr)

    # Set lưu index dòng cần xóa
    bad_indexes = set()

    for i, line in enumerate(dataset):
        words = set(line.split())  # tạo set từ dòng
        if words & target_words:   # nếu có giao nhau
            bad_indexes.add(i)

    return bad_indexes
```
