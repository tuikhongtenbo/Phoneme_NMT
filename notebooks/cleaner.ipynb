```python
pattern = r"\b(?=\w*\d)(?=\w*[A-Za-z])\w+\b"

for i in range(len(lines_all)):
    while True:
        matches = re.findall(pattern, lines_all[i])
        if not matches:
            break
        for word in matches:
            parts = re.findall(r'\d+|[A-Za-z]+', word)
            lines_all[i] = lines_all[i].replace(word, ' '.join(parts))
```

```python
def extract_vocab_from_sentences(sentences):
    vocab = set()
    for sen in sentences: 
        sen = re.sub(r"\s+", " ", sen).strip() 

        tokens = sen.split()

        tokens = [re.sub(r"^'+|'+$", "", tok) for tok in tokens if tok.strip("'")]
        vocab.update(tokens)

    return sorted(vocab)

vocab = extract_vocab_from_sentences([lines for lines in lines_all])
```

```python
import re

pairs = [
    (r"C:\Users\ply58\OneDrive\Tài liệu\PythonML\Research\MaTrans_En_Vie\uit-eng-vi-translation\Phoneme_NMT\dataset\vocabs\clean\train_end.en",
     r"C:\Users\ply58\OneDrive\Tài liệu\PythonML\Research\MaTrans_En_Vie\uit-eng-vi-translation\Phoneme_NMT\dataset\vocabs\clean\train_end.vi"),
    (r"C:\Users\ply58\OneDrive\Tài liệu\PythonML\Research\MaTrans_En_Vie\uit-eng-vi-translation\Phoneme_NMT\dataset\vocabs\clean\dev_end.en",
     r"C:\Users\ply58\OneDrive\Tài liệu\PythonML\Research\MaTrans_En_Vie\uit-eng-vi-translation\Phoneme_NMT\dataset\vocabs\clean\dev_end.vi"),
     (r"C:\Users\ply58\OneDrive\Tài liệu\PythonML\Research\MaTrans_En_Vie\uit-eng-vi-translation\Phoneme_NMT\dataset\vocabs\clean\test_end.en",
     r"C:\Users\ply58\OneDrive\Tài liệu\PythonML\Research\MaTrans_En_Vie\uit-eng-vi-translation\Phoneme_NMT\dataset\vocabs\clean\test_end.vi")
]

def clean_pairs(en_path, vi_path):
    with open(en_path, 'r', encoding='utf-8') as f_en:
        en_lines = f_en.readlines()
    with open(vi_path, 'r', encoding='utf-8') as f_vi:
        vi_lines = f_vi.readlines()
    
    assert len(en_lines) == len(vi_lines), f"File length mismatch: {en_path} & {vi_path}"

    pattern = re.compile(r'(http|https|www)', re.IGNORECASE)

    # Lines to del
    remove_indices = [i for i, line in enumerate(en_lines) if pattern.search(line)]
    print(f"\n{en_path.split('\\')[-1]}: {len(remove_indices)} lines will be removed")
    for idx in remove_indices[:5]:
        print(f"  Line {idx}: {en_lines[idx].strip()}")

    # Create clean version
    en_clean = [line for i, line in enumerate(en_lines) if i not in remove_indices]
    vi_clean = [line for i, line in enumerate(vi_lines) if i not in remove_indices]

    print(f"  Before: {len(en_lines)} lines → After: {len(en_clean)} lines")

    # Write new files
    with open(en_path.replace('.en', '_clean.en'), 'w', encoding='utf-8') as f_en:
        f_en.writelines(en_clean)
    with open(vi_path.replace('.vi', '_clean.vi'), 'w', encoding='utf-8') as f_vi:
        f_vi.writelines(vi_clean)

    return remove_indices
```

```python
for en_file, vi_file in pairs:
    clean_pairs(en_file, vi_file)
```
